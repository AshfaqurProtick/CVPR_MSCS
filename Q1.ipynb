```python
a= [5,6,80,2,10,41,21,345,53,8,71,1,42,32]
```

# Insertion Sort

```python
def insertionSort(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i-1
        while j >=0 and key < a[j] :
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
        print(f'Interation {i}--->{a}')

insertionSort(a)
```

# Quick Sort

```python
def quicksort(x):
    if len(x) == 1 or len(x) == 0:
        return x
    else:
        pivot = x[0]
        i = 0
        for j in range(len(x)-1):
            if x[j+1] < pivot:
                x[j+1],x[i+1] = x[i+1], x[j+1]
                i += 1
            print(f'Step: {j}-->{x}')
        print('---- next recursion ----')
        x[0],x[i] = x[i],x[0]
        first_part = quicksort(x[:i])
        second_part = quicksort(x[i+1:])
        first_part.append(x[i])
        return first_part + second_part
        
quicksort(a)
print(a)
```
